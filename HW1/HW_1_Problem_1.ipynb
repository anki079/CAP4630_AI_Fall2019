```
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np
from mpl_toolkits import mplot3d
import math
```

**Function 1: a local maximum**

```
fig = plt.figure()
ax = plt.axes(projection = '3d')

def f(x,y):
  return -(2 * x ** 2) - (9 * y ** 2) + 5
 
x = np.linspace(-5, 5, 30)
y = np.linspace(-5, 5, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

ax.contour3D(X, Y, Z, 50, cmap='YlGn')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');

plt.plot([0],[0],[0],markerfacecolor= 'r', markeredgecolor = 'b', marker = 'o', markersize = 10, zorder=100)

```
